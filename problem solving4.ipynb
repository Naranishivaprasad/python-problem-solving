```python
# Hollow Butterfly Pattern
# Input: n = 4
n = 4
for i in range(n, 0, -1):
    s = " " * (n - i)
    if i == 1:
        print("*" + s + "*" + s + "*")
    else:
        if i == n:
            print("* " + "  " * (i - 2) + "  *")
            print("*" + s + "* " + "  " * (i - 2) + "*" + s + "*")
        else:
            print("*" + s + "* " + "  " * (i - 2) + "*" + s + "*")

for i in range(2, n + 1):
    s = " " * (n - i)
    if i == n:
        print("*" + s + "* " + "  " * (i - 2) + "*" + s + "*")
        print("* " + "  " * (i - 2) + "  *")
    else:
        print("*" + s + "* " + "  " * (i - 2) + "*" + s + "*")
```

```python
# Hollow Hourglass Pattern
# Input: n = 5

n=4
for i in range(n-1,-1,-1):
    for k in range(n-i):
        print(" ",end=' ')
    for j in range(2*i+1):
        if i==n-1 or j==0 or j==2*i:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
for i in range(1,n):
    for k in range(n-i):
        print(" ",end=" ")
    for j in range(2*i+1):
        if i==n-1 or j==0 or j==2*i:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```
