```python
# 1. Print All Prime Numbers from m to n
# Problem: Given a range from m to n, print all prime numbers in that range.
# Input: m = 10, n = 30
def prime(n):
    if n < 2:
        return False
    for i in range(2, n):  
        if n % i == 0:
            return False
    return True

m = 10
n = 30

for i in range(m, n + 1):
    if prime(i):
        print(i,end=" ") 
```

```python
# 2. Count of All Prime Numbers from m to n
# Problem: Count how many prime numbers are there between m and n.
# Input: m = 1, n = 10

def prime(n):
    if n < 2:
        return False
    for i in range(2, n):  
        if n % i == 0:
            return False
    return True
m = 1
n = 10
count=0

for i in range(m, n + 1):
    if prime(i):
        count+=1 
print(count)
```

```python
# 3. Print All Armstrong Numbers in a Range
# Problem: Print all Armstrong numbers between m and n.
# Input: m = 1, n = 500
# Output: 1 153 370 371 407
m = 1
n = 500

for i in range(m, n + 1):
    count = 0
    res = 0
    a = i
    while a > 0:
        count += 1
        a //= 10
    b = i
    while b > 0:
        r = b % 10
        res += r ** count
        b //= 10
    if   res == i:
        print(i,end=" ")
```

```python
# 5. Last Prime Number from m to n
# Problem: Find the last prime number in the given range.
# Input: m = 10, n = 25
# Output: 23
def prime(n):
    if n < 2:
        return False
    for i in range(2, n):  
        if n % i == 0:
            return False
    return True
m = 10
n = 25
last=None

for i in range(m, n + 1):
    if prime(i):
        last=i
print(last)
```

```python
# 10. Count of Prime and Composite Numbers from m to n
# Problem: Count how many are prime and how many are composite numbers in range m to n.
# Input: m = 1, n = 10
# Output: Prime: 4, Composite: 4
m=1
n=10
prime=0
compos=0
for i in range(m,n):
    if i<=1:
        continue
    for j in range(2,i):
        if i%j==0:
            compos+=1
            break
    else:
        prime+=1
print(prime,compos)
```
